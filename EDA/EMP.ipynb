```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculer la matrice de corrÃ©lation
correlation_matrix = df.corr()

# Afficher la matrice de corrÃ©lation sous forme de heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Afficher le graphique
plt.show()
```

```python
from numpy import median
from numpy import abs

# Calculer la mÃ©diane et la MAD
median_marital = df['MARITAL_STATUS'].median()
mad_marital = abs(df['MARITAL_STATUS'] - median_marital).median()

# DÃ©finir une limite pour identifier les valeurs aberrantes
outlier_limit = 3 * mad_marital

# Identifier les valeurs aberrantes
outliers_marital = df[abs(df['MARITAL_STATUS'] - median_marital) > outlier_limit]
```
