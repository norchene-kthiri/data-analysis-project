```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix on numeric columns
correlation_matrix = numeric_df.corr()

# Display the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Show the plot
plt.show()
```

```python
from numpy import median
from numpy import abs

# Calculate the median and MAD
median_marital = df['MARITAL_STATUS'].median()
mad_marital = abs(df['MARITAL_STATUS'] - median_marital).median()

# Define a threshold to identify outliers
outlier_limit = 3 * mad_marital

# Identify outliers
outliers_marital = df[abs(df['MARITAL_STATUS'] - median_marital) > outlier_limit]
```

```python
# Group the data by gender and count the frequency of technologies for each group
technologies_by_gender = df.groupby('GENDER')['TECHNOLOGIES'].value_counts()
print(technologies_by_gender)
```

```python
# Group the data by gender and count the frequency of technologies for each group
technologies_by_gender = df.groupby('GENDER')['TECHNOLOGIES'].value_counts().unstack()

# Display a stacked bar chart
ax = technologies_by_gender.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title('Technologies by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.legend(title='Technologies', bbox_to_anchor=(1, 1))

# Annotate each bar with the corresponding values
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{int(height)}', (x + width/2, y + height/2), ha='center', va='center')

plt.show()
```
