```python
# Bivariate Analysis - Numerical vs Categorical (CAPACITY vs DEP)

# Aggregate capacity values by department name
capacity_by_dep = df.groupby('DEP')['CAPACITY'].sum().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='DEP', y='CAPACITY', data=capacity_by_dep, palette='pastel')
plt.title('Total Capacity by Department')
plt.xlabel('Department Name')
plt.ylabel('Total Capacity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bivariate Analysis - Numerical vs Categorical (CAPACITY vs DEP)

# Aggregate capacity values by department name
capacity_by_dep = df.groupby('DEP')['CAPACITY'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(capacity_by_dep, labels=capacity_by_dep.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Department Capacities')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
