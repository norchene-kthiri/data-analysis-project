# Objective 2 :  Contract and Tenure Analysis:

```python
#Pour avoir une idÃ©e sur les types des contrats
TYPE_CONTRACT_counts = administration_data['TYPE_CONTRACT'].value_counts()

print(TYPE_CONTRACT_counts)
```

```python
# 1. Analyse de la rÃ©partition des types de contrat
contract_type_distribution = administration_data['TYPE_CONTRACT'].value_counts()

# CrÃ©er un diagramme Ã  barres ou un pie chart pour la rÃ©partition des types de contrat
plt.figure(figsize=(8, 6))
contract_type_distribution.plot(kind='bar', color='skyblue')
plt.title('RÃ©partition des Types de Contrat')
plt.xlabel('Type de Contrat')
plt.ylabel('Nombre d\'EmployÃ©s')
plt.show()
```

```python
#Afin d'obtenir un aperÃ§u du nombre d'employÃ©s dont le contrat est commencÃ© pour chaque annÃ©e.


# Convertir la colonne "CONTRACT_END" en format de date si elle ne l'est pas dÃ©jÃ 
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')

# Filtrer les lignes oÃ¹ le contrat s'est terminÃ© entre 2021 et 2025
terminated_between_2021_2025 = administration_data[(administration_data['CONTRACT_START'].dt.year >= 2010) & (administration_data['CONTRACT_START'].dt.year <= 2025)]

# Utiliser groupby pour obtenir le dÃ©compte pour chaque annÃ©e
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_START'].dt.year).size()

# Afficher le dÃ©compte pour chaque annÃ©e
print(terminated_by_year)


# 2. Analyse temporelle des dates de dÃ©but de contrat
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')


# CrÃ©er un graphique pour la distribution des dates de dÃ©but de contrat
plt.figure(figsize=(12, 6))
plt.hist(administration_data['CONTRACT_START'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution des Dates de DÃ©but de Contrat')
plt.xlabel('Date de DÃ©but de Contrat')
plt.ylabel('Nombre d\'EmployÃ©s')
plt.show()

```

```python
# Afin d'obtenir un aperÃ§u du nombre d'employÃ©s dont le contrat s'est terminÃ© pour chaque annÃ©e.


# Convertir la colonne "CONTRACT_END" en format de date si elle ne l'est pas dÃ©jÃ 
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')

# Filtrer les lignes oÃ¹ le contrat s'est terminÃ© entre 2021 et 2025
terminated_between_2021_2025 = administration_data[(administration_data['CONTRACT_END'].dt.year >= 2021) & (administration_data['CONTRACT_END'].dt.year <= 2025)]

# Utiliser groupby pour obtenir le dÃ©compte pour chaque annÃ©e
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_END'].dt.year).size()

# Afficher le dÃ©compte pour chaque annÃ©e
print(terminated_by_year)

```

```python

# CrÃ©er un graphique pour la distribution des dates de fin de contrat
plt.figure(figsize=(12, 6))
plt.hist(administration_data['CONTRACT_END'].dropna(), bins=20, color='lightcoral', edgecolor='black')
plt.title('Distribution des Dates de Fin de Contrat')
plt.xlabel('Date de Fin de Contrat')
plt.ylabel('Nombre d\'EmployÃ©s')
plt.show()
```

```python
# 3. Analyse de la durÃ©e de service
administration_data['SERVICE_DURATION'] = administration_data['CONTRACT_END'] - administration_data['CONTRACT_START']
service_duration_stats = administration_data['SERVICE_DURATION'].describe()

# Afficher les statistiques descriptives de la durÃ©e de service
print("Statistiques descriptives de la durÃ©e de service des employÃ©s :\n", service_duration_stats)

# CrÃ©er un histogramme pour la distribution de la durÃ©e de service
plt.figure(figsize=(10, 6))
plt.hist(administration_data['SERVICE_DURATION'].dropna().dt.days, bins=20, color='orange', edgecolor='black')
plt.title('Distribution de la DurÃ©e de Service des EmployÃ©s')
plt.xlabel('DurÃ©e de Service (jours)')
plt.ylabel('Nombre d\'EmployÃ©s')
plt.show()
```
