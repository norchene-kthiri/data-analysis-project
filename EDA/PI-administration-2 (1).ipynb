```python
# 6. Analyse croisée : Distribution des grades par genre
cross_gender_grade = pd.crosstab(administration_data['GENDER'], administration_data['Grade'])

# Créer un diagramme à barres empilées pour la distribution des grades par genre
ax = cross_gender_grade.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Distribution des Grades par Genre')
plt.xlabel('Genre')
plt.ylabel('Nombre d\'Employés')

# Ajouter les valeurs sur les barres empilées
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width / 2, y + height / 2), ha='center', va='center', fontsize=10)

plt.show()
```

```python
# 7. Analyse temporelle : Distribution des employés par année d'embauche
administration_data['STARTING_DATE'] = pd.to_datetime(administration_data['STARTING_DATE'], errors='coerce')
starting_year_distribution = administration_data['STARTING_DATE'].dt.year.value_counts().sort_index()

# Créer un diagramme à barres pour la distribution des employés par année d'embauche
plt.figure(figsize=(12, 6))
starting_year_distribution.plot(kind='bar', color='purple')
plt.title('Distribution des Employés par Année d\'Embauche')
plt.xlabel('Année d\'Embauche')
plt.ylabel('Nombre d\'Employés')
plt.show()
```

```python
# Afin d'obtenir une vue sur les grades ayant le type de contrat "CDD", facilitant la gestion des effectifs.
cdd_counts = administration_data[administration_data['TYPE_CONTRACT'] == 'CDD'].groupby('Grade').size()

print(cdd_counts)

```

```python
# Afin d'obtenir une vue sur le nombre d'employés par grade ayant une valeur NULL dans la colonne "CONTRACT_END".
null_contract_end_counts = administration_data[administration_data['CONTRACT_END'].isnull()].groupby('Grade').size()
print(null_contract_end_counts)
```

```python
# Supposons que votre DataFrame s'appelle administration_data
# Filtrer les données pour ne conserver que celles avec le grade "RH"
rh_data = administration_data[administration_data['Grade'] == 'Financier']

# Tri des données par salaire en ordre décroissant
top_rh_salaries = rh_data.sort_values(by='SALARY', ascending=False).head(5)

# Création du graphique à barres
plt.figure(figsize=(12, 6))
bars = plt.bar(top_rh_salaries['NAME_EMP'], top_rh_salaries['SALARY'])

# Ajouter les annotations de salaire sur chaque barre
for bar, salary in zip(bars, top_rh_salaries['SALARY']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{salary:.2f}', ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Salaries for Employees with Grade "Financier"', fontsize=15)
plt.xlabel('Employee Names', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.ylabel('Salary', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
```

# Objective 2 :  Contract and Tenure Analysis:

```python
#Pour avoir une idée sur les types des contrats
TYPE_CONTRACT_counts = administration_data['TYPE_CONTRACT'].value_counts()

print(TYPE_CONTRACT_counts)
```

```python
# 1. Analyse de la répartition des types de contrat
contract_type_distribution = administration_data['TYPE_CONTRACT'].value_counts()

# Créer un diagramme à barres ou un pie chart pour la répartition des types de contrat
plt.figure(figsize=(8, 6))
contract_type_distribution.plot(kind='bar', color='skyblue')
plt.title('Répartition des Types de Contrat')
plt.xlabel('Type de Contrat')
plt.ylabel('Nombre d\'Employés')
plt.show()
```

```python
#Afin d'obtenir un aperçu du nombre d'employés dont le contrat est commencé pour chaque année.


# Convertir la colonne "CONTRACT_END" en format de date si elle ne l'est pas déjà
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')

# Filtrer les lignes où le contrat s'est terminé entre 2021 et 2025
terminated_between_2021_2025 = administration_data[(administration_data['CONTRACT_START'].dt.year >= 2010) & (administration_data['CONTRACT_START'].dt.year <= 2025)]

# Utiliser groupby pour obtenir le décompte pour chaque année
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_START'].dt.year).size()

# Afficher le décompte pour chaque année
print(terminated_by_year)


# 2. Analyse temporelle des dates de début de contrat
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')


# Créer un graphique pour la distribution des dates de début de contrat
plt.figure(figsize=(12, 6))
plt.hist(administration_data['CONTRACT_START'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution des Dates de Début de Contrat')
plt.xlabel('Date de Début de Contrat')
plt.ylabel('Nombre d\'Employés')
plt.show()

```

```python
# Afin d'obtenir un aperçu du nombre d'employés dont le contrat s'est terminé pour chaque année.


# Convertir la colonne "CONTRACT_END" en format de date si elle ne l'est pas déjà
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')

# Filtrer les lignes où le contrat s'est terminé entre 2021 et 2025
terminated_between_2021_2025 = administration_data[(administration_data['CONTRACT_END'].dt.year >= 2021) & (administration_data['CONTRACT_END'].dt.year <= 2025)]

# Utiliser groupby pour obtenir le décompte pour chaque année
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_END'].dt.year).size()

# Afficher le décompte pour chaque année
print(terminated_by_year)

```

```python

# Créer un graphique pour la distribution des dates de fin de contrat
plt.figure(figsize=(12, 6))
plt.hist(administration_data['CONTRACT_END'].dropna(), bins=20, color='lightcoral', edgecolor='black')
plt.title('Distribution des Dates de Fin de Contrat')
plt.xlabel('Date de Fin de Contrat')
plt.ylabel('Nombre d\'Employés')
plt.show()
```

```python
# 3. Analyse de la durée de service
administration_data['SERVICE_DURATION'] = administration_data['CONTRACT_END'] - administration_data['CONTRACT_START']
service_duration_stats = administration_data['SERVICE_DURATION'].describe()

# Afficher les statistiques descriptives de la durée de service
print("Statistiques descriptives de la durée de service des employés :\n", service_duration_stats)

# Créer un histogramme pour la distribution de la durée de service
plt.figure(figsize=(10, 6))
plt.hist(administration_data['SERVICE_DURATION'].dropna().dt.days, bins=20, color='orange', edgecolor='black')
plt.title('Distribution de la Durée de Service des Employés')
plt.xlabel('Durée de Service (jours)')
plt.ylabel('Nombre d\'Employés')
plt.show()
```
