```python
# 6. Cross-analysis: Distribution of grades by gender
cross_gender_grade = pd.crosstab(administration_data['GENDER'], administration_data['Grade'])

# Create a stacked bar chart for the distribution of grades by gender
ax = cross_gender_grade.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Distribution of Grades by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')

# Add values on the stacked bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width / 2, y + height / 2), ha='center', va='center', fontsize=10)

plt.show()

```

```python
# 7. Time analysis: Distribution of employees by year of hiring
administration_data['STARTING_DATE'] = pd.to_datetime(administration_data['STARTING_DATE'], errors='coerce')
starting_year_distribution = administration_data['STARTING_DATE'].dt.year.value_counts().sort_index()

# Create a bar chart for the distribution of employees by year of hiring
plt.figure(figsize=(12, 6))
starting_year_distribution.plot(kind='bar', color='purple')
plt.title('Distribution of Employees by Year of Hiring')
plt.xlabel('Year of Hiring')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Assuming your DataFrame is named administration_data
# Filter data to keep only those with grade "Financier"
rh_data = administration_data[administration_data['Grade'] == 'Financier']

# Sort data by salary in descending order
top_rh_salaries = rh_data.sort_values(by='SALARY', ascending=False).head(5)

# Create bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(top_rh_salaries['NAME_EMP'], top_rh_salaries['SALARY'])

# Add salary annotations on each bar
for bar, salary in zip(bars, top_rh_salaries['SALARY']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{salary:.2f}', ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Salaries for Employees with Grade "Financier"', fontsize=15)
plt.xlabel('Employee Names', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.ylabel('Salary', fontsize=12)
plt.yticks(fontsize=10)
plt.show()

```

# Objective 2 :  Contract and Tenure Analysis:

```python
# To get an idea of the types of contracts
TYPE_CONTRACT_counts = administration_data['TYPE_CONTRACT'].value_counts()

print(TYPE_CONTRACT_counts)
```

```python
# 1. Analysis of contract type distribution
contract_type_distribution = administration_data['TYPE_CONTRACT'].value_counts()

# Create a pie chart for the distribution of contract types
plt.figure(figsize=(8, 6))
contract_type_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Contract Types')
plt.ylabel('')  
plt.show()
```

```python
# To get an overview of the number of employees whose contracts started each year.

# Convert the "CONTRACT_START" column to datetime format if not already
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')

# Filter rows where contract started between 2010 and 2025
started_between_2010_2025 = administration_data[
    (administration_data['CONTRACT_START'].dt.year >= 2010) & 
    (administration_data['CONTRACT_START'].dt.year <= 2025)
]

# Use groupby to get the count for each year
started_by_year = started_between_2010_2025.groupby(started_between_2010_2025['CONTRACT_START'].dt.year).size()

# Display the count for each year
print(started_by_year)


# 2. Temporal analysis of contract start dates
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')

# Create a histogram for the distribution of contract start dates
plt.figure(figsize=(12, 6))
plt.hist(administration_data['CONTRACT_START'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Contract Start Dates')
plt.xlabel('Contract Start Date')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# To get an overview of the number of employees whose contracts ended each year.

# Convert the "CONTRACT_END" column to datetime format if it isn't already
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')

# Filter rows where the contract ended between 2021 and 2025
terminated_between_2021_2025 = administration_data[
    (administration_data['CONTRACT_END'].dt.year >= 2021) & 
    (administration_data['CONTRACT_END'].dt.year <= 2025)
]

# Use groupby to get the count for each year
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_END'].dt.year).size()

# Display the count for each year
print(terminated_by_year)


```

```python
# Create a histogram for the distribution of contract end dates
plt.figure(figsize=(12, 6))
plt.hist(administration_data['CONTRACT_END'].dropna(), bins=20, color='lightcoral', edgecolor='black')
plt.title('Distribution of Contract End Dates')
plt.xlabel('Contract End Date')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# 3. Analysis of Service Duration
administration_data['SERVICE_DURATION'] = administration_data['CONTRACT_END'] - administration_data['CONTRACT_START']
service_duration_stats = administration_data['SERVICE_DURATION'].describe()

# Display descriptive statistics of service duration
print("Descriptive statistics of employees' service duration:\n", service_duration_stats)

# Create a histogram for the distribution of service duration
plt.figure(figsize=(10, 6))
plt.hist(administration_data['SERVICE_DURATION'].dropna().dt.days, bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Employees\' Service Duration')
plt.xlabel('Service Duration (days)')
plt.ylabel('Number of Employees')
plt.show()
```

```python
salary_comparison = administration_data.groupby('Grade')['SALARY'].mean().sort_values()

# Create a bar plot to compare average salaries across different grades
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x=salary_comparison.index, y=salary_comparison.values, palette='viridis')

# Add the salary value above each bar
for index, value in enumerate(salary_comparison.values):
    barplot.text(index, value + 50, f'{value:.2f}', ha='center', va='bottom', fontsize=9, color='black')

plt.title('Salary Comparison by Grade')
plt.xlabel('Grade')
plt.ylabel('Average Salary')
plt.show()
```
