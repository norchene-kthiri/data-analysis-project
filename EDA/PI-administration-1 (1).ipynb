# Objective 2 :  Contract and Tenure Analysis:

```python
#Pour avoir une idÃ©e sur les types des contrats
TYPE_CONTRACT_counts = administration_data['TYPE_CONTRACT'].value_counts()

print(TYPE_CONTRACT_counts)
```

```python
# 1. Analyse de la rÃ©partition des types de contrat
contract_type_distribution = administration_data['TYPE_CONTRACT'].value_counts()

# CrÃ©er un diagramme Ã  barres pour la rÃ©partition des types de contrat
plt.figure(figsize=(8, 6))
contract_type_distribution.plot(kind='bar', color='skyblue')
plt.title('RÃ©partition des Types de Contrat')
plt.xlabel('Type de Contrat')
plt.ylabel('Nombre d\'EmployÃ©s')
plt.show()
```

```python
#Afin d'obtenir un aperÃ§u du nombre d'employÃ©s dont le contrat est commencÃ© pour chaque annÃ©e.


# Convertir la colonne "CONTRACT_END" en format de date 
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')

# Filtrer les lignes oÃ¹ le contrat s'est terminÃ© entre 2010 et 2025
terminated_between_2010_2025 = administration_data[(administration_data['CONTRACT_START'].dt.year >= 2010) & (administration_data['CONTRACT_START'].dt.year <= 2025)]

# On a utilisÃ© groupby pour obtenir le dÃ©compte pour chaque annÃ©e
terminated_by_year = terminated_between_2010_2025.groupby(terminated_between_2010_2025['CONTRACT_START'].dt.year).size()

# Afficher le dÃ©compte pour chaque annÃ©e
print(terminated_by_year)


```

```python
# Afin d'obtenir un aperÃ§u du nombre d'employÃ©s dont le contrat s'est terminÃ© pour chaque annÃ©e.


# Convertir la colonne "CONTRACT_END" en format de date 
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')

# Filtrer les lignes oÃ¹ le contrat s'est terminÃ© entre 2021 et 2025
terminated_between_2021_2025 = administration_data[(administration_data['CONTRACT_END'].dt.year >= 2021) & (administration_data['CONTRACT_END'].dt.year <= 2025)]

# on a utilisÃ© groupby pour obtenir le dÃ©compte pour chaque annÃ©e
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_END'].dt.year).size()

# Afficher le dÃ©compte pour chaque annÃ©e
print(terminated_by_year)

```

```python
# 3. Analyse de la durÃ©e de service
administration_data['SERVICE_DURATION'] = administration_data['CONTRACT_END'] - administration_data['CONTRACT_START']
service_duration_stats = administration_data['SERVICE_DURATION'].describe()

# Afficher les statistiques descriptives de la durÃ©e de service
print("Statistiques descriptives de la durÃ©e de service des employÃ©s :\n", service_duration_stats)

```

```python
# Afin d'obtenir un aperÃ§u du nombre d'heures supplÃ©mentaires par grade.
extra_hours_by_grade = administration_data.groupby('Grade')['extra_hour'].sum()

# Trier les donnÃ©es en ordre croissant
extra_hours_by_grade = extra_hours_by_grade.sort_values(ascending=True)

# CrÃ©er un graphique Ã  barres
plt.figure(figsize=(12, 6))
bars = extra_hours_by_grade.plot(kind='bar', color='orange')


plt.title('Nombre d\'Heures SupplÃ©mentaires par Grade (Tri Croissant)', fontsize=15)
plt.xlabel('Grade', fontsize=12)
plt.ylabel('Heures SupplÃ©mentaires', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.show()
```

```python
salary_comparison = administration_data.groupby('Grade')['SALARY'].mean().sort_values()

# CrÃ©er un diagramme Ã  barres pour comparer les salaires entre les diffÃ©rents grades
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x=salary_comparison.index, y=salary_comparison.values, palette='viridis')

# Pour ajouter chaque valeur de salaire au-dessus de la barre correspondante
for index, value in enumerate(salary_comparison.values):
    barplot.text(index, value + 50, f'{value:.2f}', ha='center', va='bottom', fontsize=9, color='black')

plt.title('Comparaison des Salaires par Grade')
plt.xlabel('Grade')
plt.ylabel('Salaire Moyen')
plt.show()
```
