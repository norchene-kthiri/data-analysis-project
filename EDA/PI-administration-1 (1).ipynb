# Objective 1 :  Analysis of Employee Demographics:


```python
# To get an idea about the marital status of our employees
MARITAL_STATUS_counts = administration_data['MARITAL_STATUS'].value_counts()

print(MARITAL_STATUS_counts)
```

```python
# 2. Analysis of distribution by marital status
marital_status_distribution = administration_data['MARITAL_STATUS'].value_counts()

# Create a bar chart for distribution by marital status
plt.figure(figsize=(8, 6))
marital_status_distribution.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# To get an idea of the count of each gender in our company
GENDER_counts = administration_data['GENDER'].value_counts()
print(GENDER_counts)
```

```python
# 1. Analysis of distribution by gender
gender_distribution = administration_data['GENDER'].value_counts()

# Create a bar chart for gender distribution
plt.figure(figsize=(8, 6))
gender_distribution.plot(kind='bar', color=['blue', 'pink'])
plt.title('Employee Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# To get an idea about the positions held by our employees
Grade_counts = administration_data['Grade'].value_counts()
print(Grade_counts)
```

```python
# 3. Analysis of distribution by grade
grade_distribution = administration_data['Grade'].value_counts()

# Create a horizontal bar chart for distribution by grade
plt.figure(figsize=(12, 6))
grade_distribution.plot(kind='barh', color='green')
plt.title('Employee Distribution by Grade')
plt.xlabel('Grade')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# To get an idea about the diploma levels of our employees
diploma_counts = administration_data['TYPE_DIPLOMA'].value_counts()

print(diploma_counts)
```

```python
# 4. Analysis of distribution by diploma type
diploma_distribution = administration_data['TYPE_DIPLOMA'].value_counts()

# Create a bar chart for distribution by diploma type
plt.figure(figsize=(10, 6))
diploma_distribution.plot(kind='bar', color='orange')
plt.title('Employee Distribution by Diploma Type')
plt.xlabel('Diploma Type')
plt.ylabel('Number of Employees')
plt.show()
```

```python
TENURE_counts = administration_data['TENURE'].value_counts()
print(TENURE_counts)
```

```python
# 6. Cross-analysis: Distribution of grades by gender
cross_gender_grade = pd.crosstab(administration_data['GENDER'], administration_data['Grade'])

# Create a stacked bar chart for grade distribution by gender
ax = cross_gender_grade.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Distribution of Grades by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')

# Add values on the stacked bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{int(height)}', (x + width / 2, y + height / 2), ha='center', va='center', fontsize=10)

plt.show()
```

```python
# 7. Time analysis: Distribution of employees by year of hire
administration_data['STARTING_DATE'] = pd.to_datetime(administration_data['STARTING_DATE'], errors='coerce')
starting_year_distribution = administration_data['STARTING_DATE'].dt.year.value_counts().sort_index()

# Create a bar chart for the distribution of employees by year of hire
plt.figure(figsize=(12, 6))
ax = starting_year_distribution.plot(kind='bar', color='purple')
plt.title('Employee Distribution by Year of Hire')
plt.xlabel('Year of Hire')
plt.ylabel('Number of Employees')

# Add values on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{int(height)}', (x + width / 2, y + height), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# To get an overview of the grades with the contract type "CDD", facilitating workforce management
cdd_counts = administration_data[administration_data['TYPE_CONTRACT'] == 'CDD'].groupby('Grade').size()

print(cdd_counts)
```

```python
# To get an overview of the number of employees by grade with a NULL value in the "CONTRACT_END" column
null_contract_end_counts = administration_data[administration_data['CONTRACT_END'].isnull()].groupby('Grade').size()
print(null_contract_end_counts)
```

```python
# Filter data to keep only those with grade "Financier"
rh_data = administration_data[administration_data['Grade'] == 'Financier']

# Sort data by salary in descending order and get top 5
top_rh_salaries = rh_data.sort_values(by='SALARY', ascending=False).head(5)

# Create the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(top_rh_salaries['NAME_EMP'], top_rh_salaries['SALARY'])

# Add salary annotations on each bar
for bar, salary in zip(bars, top_rh_salaries['SALARY']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{salary:.2f}', ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Salaries for Employees', fontsize=15)
plt.xlabel('Employee Names', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.ylabel('Salary', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
```

# Objective 2 :  Contract and Tenure Analysis:

```python
# To get an overview of the contract types
TYPE_CONTRACT_counts = administration_data['TYPE_CONTRACT'].value_counts()

print(TYPE_CONTRACT_counts)
```

```python
# 1. Analysis of contract type distribution
contract_type_distribution = administration_data['TYPE_CONTRACT'].value_counts()

# Create a bar chart for contract type distribution
plt.figure(figsize=(8, 6))
contract_type_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Contract Types')
plt.xlabel('Contract Type')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# To get an overview of the number of employees whose contracts started each year

# Convert the "CONTRACT_START" column to datetime format
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')

# Filter rows where the contract started between 2010 and 2025
started_between_2010_2025 = administration_data[
    (administration_data['CONTRACT_START'].dt.year >= 2010) & 
    (administration_data['CONTRACT_START'].dt.year <= 2025)
]

# Use groupby to get the count for each year
started_by_year = started_between_2010_2025.groupby(started_between_2010_2025['CONTRACT_START'].dt.year).size()

# Display the count per year
print(started_by_year)
```

```python
# To get an overview of the number of employees whose contracts ended each year

# Convert the "CONTRACT_END" column to datetime format
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')

# Filter rows where the contract ended between 2021 and 2025
terminated_between_2021_2025 = administration_data[
    (administration_data['CONTRACT_END'].dt.year >= 2021) & 
    (administration_data['CONTRACT_END'].dt.year <= 2025)
]

# Use groupby to get the count for each year
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_END'].dt.year).size()

# Display the count per year
print(terminated_by_year)
```

```python
# 3. Analysis of length of service
administration_data['SERVICE_DURATION'] = administration_data['CONTRACT_END'] - administration_data['CONTRACT_START']
service_duration_stats = administration_data['SERVICE_DURATION'].describe()

# Display descriptive statistics of length of service
print("Descriptive statistics of employees' length of service:\n", service_duration_stats)
```

# objective 5 :   Compensation and Overtime Analysis:


```python
# To get an overview of the number of overtime hours by grade
extra_hours_by_grade = administration_data.groupby('Grade')['extra_hour'].sum()

# Sort the data in ascending order
extra_hours_by_grade = extra_hours_by_grade.sort_values(ascending=True)

# Create a bar chart
plt.figure(figsize=(12, 6))
extra_hours_by_grade.plot(kind='bar', color='orange')

plt.title('Overtime Hours by Grade (Ascending Order)', fontsize=15)
plt.xlabel('Grade', fontsize=12)
plt.ylabel('Overtime Hours', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.show()
```

```python
# To get an overview of the number of overtime hours by diploma type
extra_hours_by_diploma = administration_data.groupby('TYPE_DIPLOMA')['extra_hour'].sum()

# Create a bar chart
plt.figure(figsize=(12, 6))
extra_hours_by_diploma.plot(kind='bar', color='orange')
plt.title('Overtime Hours by Diploma Type', fontsize=15)
plt.xlabel('Diploma Type', fontsize=12)
plt.ylabel('Overtime Hours', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.show()


```
