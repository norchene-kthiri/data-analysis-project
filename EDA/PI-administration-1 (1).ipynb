Il semble y avoir un mélange de données dans votre exemple, mais je vais expliquer l'utilité de chaque colonne en supposant que les données sont structurées comme suit :

1. *CODE :* Identifiant unique associé à chaque employé (par exemple, numéro d'identification unique).

2. *NAME_EMP :* Le nom de l'employé (dans cet exemple, Tony Freeman).

3. *GENDER :* Le genre de l'employé (F pour féminin, M pour masculin, etc.).

4. *ADDRESS :* L'adresse de résidence de l'employé.

5. *MARITAL_STATUS :* Le statut matrimonial de l'employé (0 pour célibataire, 1 pour marié, etc.).

6. *Grade :* Le niveau ou le grade de l'employé au sein de l'organisation (par exemple, RH pour Ressources Humaines).

7. *TYPE_DIPLOMA :* Le type de diplôme détenu par l'employé (par exemple, Master).

8. *EXP_YEARS :* Le nombre d'années d'expérience professionnelle de l'employé.

9. *EMAIL :* L'adresse e-mail de l'employé.

10. *LANGUAGE :* Les langues que l'employé maîtrise.

11. *extra_hour :* Le nombre d'heures supplémentaires travaillées par l'employé.

12. *POSITION :* Le poste occupé par l'employé au sein de l'organisation.

13. *SALARY :* Le salaire de l'employé.

14. *STARTING_DATE :* La date à laquelle l'employé a commencé à travailler dans l'entreprise.

15. *REGISTRATION_NUMBER :* Un numéro d'enregistrement associé à l'employé.

16. *TYPE_CONTRACT :* Le type de contrat de travail (par exemple, CDD pour Contrat à Durée Déterminée).

17. *DEP_ID :* L'identifiant du département auquel l'employé est rattaché.

18. *LEAVE_CONSUMED :* Le nombre de jours de congé déjà pris par l'employé.

19. *LEAVE_BALANCE :* Le solde de jours de congé restant pour l'employé.

20. *CONTRACT_START :* La date de début du contrat de travail.

21. *CONTRACT_END :* La date de fin du contrat de travail.

22. *TENURE :* La durée totale du service de l'employé dans l'entreprise.   !!!!!!!!!!!!!!!!!!!!!!!!!!!!!

23. *LAST_LEAVE_TYPE :* Le type de dernier congé pris par l'employé.

24. *SOURCE_of_employment :* La source par laquelle l'employé a été recruté (par exemple, "Free Application" dans cet exemple).

Chaque colonne dans ces données contient des informations spécifiques sur l'employé, permettant une gestion et une analyse détaillées des ressources humaines au sein de l'organisation.

```python
# 6. Analyse croisée : Distribution des grades par genre
cross_gender_grade = pd.crosstab(administration_data['GENDER'], administration_data['Grade'])

# Créer un diagramme à barres empilées pour la distribution des grades par genre
ax = cross_gender_grade.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Distribution des Grades par Genre')
plt.xlabel('Genre')
plt.ylabel('Nombre d\'Employés')

# Ajouter les valeurs sur les barres empilées
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{int(height)}', (x + width / 2, y + height / 2), ha='center', va='center', fontsize=10)

plt.show()
```

```python
# 7. Analyse temporelle : Distribution des employés par année d'embauche
administration_data['STARTING_DATE'] = pd.to_datetime(administration_data['STARTING_DATE'], errors='coerce')
starting_year_distribution = administration_data['STARTING_DATE'].dt.year.value_counts().sort_index()

# Créer un diagramme à barres pour la distribution des employés par année d'embauche
plt.figure(figsize=(12, 6))
ax = starting_year_distribution.plot(kind='bar', color='purple')
plt.title('Distribution des Employés par Année d\'Embauche')
plt.xlabel('Année d\'Embauche')
plt.ylabel('Nombre d\'Employés')

# Ajouter les valeurs sur les barres
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{int(height)}', (x + width / 2, y + height), ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Filtrer les données pour ne conserver que celles avec le grade "Financier"
 #rh_data = administration_data[administration_data['Grade']]

# Tri des données par salaire en ordre décroissant
top_rh_salaries = administration_data.sort_values(by='SALARY', ascending=False).head(5)

# Création du graphique à barres
plt.figure(figsize=(12, 6))
bars = plt.bar(top_rh_salaries['NAME_EMP'], top_rh_salaries['SALARY'])

# Ajouter les annotations de salaire sur chaque barre
for bar, salary in zip(bars, top_rh_salaries['SALARY']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{salary:.2f}', ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Salaries for Employees"', fontsize=15)
plt.xlabel('Employee Names', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.ylabel('Salary', fontsize=12)
plt.yticks(fontsize=10)
plt.show()
```

# Objective 2 :  Contract and Tenure Analysis:

```python
#Pour avoir une idée sur les types des contrats
TYPE_CONTRACT_counts = administration_data['TYPE_CONTRACT'].value_counts()

print(TYPE_CONTRACT_counts)
```

```python
# 1. Analyse de la répartition des types de contrat
contract_type_distribution = administration_data['TYPE_CONTRACT'].value_counts()

# Créer un diagramme à barres pour la répartition des types de contrat
plt.figure(figsize=(8, 6))
contract_type_distribution.plot(kind='bar', color='skyblue')
plt.title('Répartition des Types de Contrat')
plt.xlabel('Type de Contrat')
plt.ylabel('Nombre d\'Employés')
plt.show()
```

```python
#Afin d'obtenir un aperçu du nombre d'employés dont le contrat est commencé pour chaque année.


# Convertir la colonne "CONTRACT_END" en format de date 
administration_data['CONTRACT_START'] = pd.to_datetime(administration_data['CONTRACT_START'], errors='coerce')

# Filtrer les lignes où le contrat s'est terminé entre 2010 et 2025
terminated_between_2010_2025 = administration_data[(administration_data['CONTRACT_START'].dt.year >= 2010) & (administration_data['CONTRACT_START'].dt.year <= 2025)]

# On a utilisé groupby pour obtenir le décompte pour chaque année
terminated_by_year = terminated_between_2010_2025.groupby(terminated_between_2010_2025['CONTRACT_START'].dt.year).size()

# Afficher le décompte pour chaque année
print(terminated_by_year)


```

```python
# Afin d'obtenir un aperçu du nombre d'employés dont le contrat s'est terminé pour chaque année.


# Convertir la colonne "CONTRACT_END" en format de date 
administration_data['CONTRACT_END'] = pd.to_datetime(administration_data['CONTRACT_END'], errors='coerce')

# Filtrer les lignes où le contrat s'est terminé entre 2021 et 2025
terminated_between_2021_2025 = administration_data[(administration_data['CONTRACT_END'].dt.year >= 2021) & (administration_data['CONTRACT_END'].dt.year <= 2025)]

# on a utilisé groupby pour obtenir le décompte pour chaque année
terminated_by_year = terminated_between_2021_2025.groupby(terminated_between_2021_2025['CONTRACT_END'].dt.year).size()

# Afficher le décompte pour chaque année
print(terminated_by_year)

```

```python
# 3. Analyse de la durée de service
administration_data['SERVICE_DURATION'] = administration_data['CONTRACT_END'] - administration_data['CONTRACT_START']
service_duration_stats = administration_data['SERVICE_DURATION'].describe()

# Afficher les statistiques descriptives de la durée de service
print("Statistiques descriptives de la durée de service des employés :\n", service_duration_stats)

```

```python
# Afin d'obtenir un aperçu du nombre d'heures supplémentaires par grade.
extra_hours_by_grade = administration_data.groupby('Grade')['extra_hour'].sum()

# Trier les données en ordre croissant
extra_hours_by_grade = extra_hours_by_grade.sort_values(ascending=True)

# Créer un graphique à barres
plt.figure(figsize=(12, 6))
bars = extra_hours_by_grade.plot(kind='bar', color='orange')


plt.title('Nombre d\'Heures Supplémentaires par Grade (Tri Croissant)', fontsize=15)
plt.xlabel('Grade', fontsize=12)
plt.ylabel('Heures Supplémentaires', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.show()
```

```python
# Afin d'obtenir un aperçu du nombre d'heures supplémentaires par grade.
extra_hours_by_grade = administration_data.groupby('TYPE_DIPLOMA')['extra_hour'].sum()

# Créer un graphique à barres
plt.figure(figsize=(12, 6))
extra_hours_by_grade.plot(kind='bar', color='orange')
plt.title('Nombre d\'Heures Supplémentaires par Grade', fontsize=15)
plt.xlabel('Grade', fontsize=12)
plt.ylabel('Heures Supplémentaires', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.show()

```

```python
salary_comparison = administration_data.groupby('Grade')['SALARY'].mean().sort_values()

# Créer un diagramme à barres pour comparer les salaires entre les différents grades
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x=salary_comparison.index, y=salary_comparison.values, palette='viridis')

# Pour ajouter chaque valeur de salaire au-dessus de la barre correspondante
for index, value in enumerate(salary_comparison.values):
    barplot.text(index, value + 50, f'{value:.2f}', ha='center', va='bottom', fontsize=9, color='black')

plt.title('Comparaison des Salaires par Grade')
plt.xlabel('Grade')
plt.ylabel('Salaire Moyen')
plt.show()
```
